```python
# Q1: Libraries + Data Load
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

df = pd.read_csv('Amazon.csv')
print("✅ File Load Ho Gayi - Project Start")
df.head()
```

```python
print("Q2: Dataset Info")
print("Total Rows, Columns:", df.shape)
print("\nColumn Names:", df.columns.tolist())
df.info()
df.describe()
```

```python
# Date column fix karo pehle
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# Is dataset me 'TotalAmount' ya 'Amount' column hai Sales ke liye
# Check kar le: df.columns me dekh 'TotalAmount' hai ya 'Amount'
total_sales = df['TotalAmount'].sum()  # Agar 'Amount' naam hai to df['Amount'].sum() karna
total_orders = df['OrderID'].nunique()
avg_order_value = df['TotalAmount'].mean()

print("Q3: KPI")
print(f"Total Sales: ₹{total_sales:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Order Value: ₹{avg_order_value:,.2f}")
```

```python
print("Q4: Top 10 States by Sales")
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False).head(10)

print("\nQ5: Sales by Category")
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False)

print("\nQ6: Sales by Payment Method")
df.groupby('PaymentMethod')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(16,6))

# Q15: Bar Chart - Top 10 States by Sales
plt.subplot(1,2,1)
df.groupby('State')['TotalAmount'].sum().nlargest(10).plot(kind='bar', color='skyblue')
plt.title('Q15: Top 10 States by Sales')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Q16: Pie Chart - Sales by Category
plt.subplot(1,2,2)
df.groupby('Category')['TotalAmount'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Q16: Sales Distribution by Category')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Q7 to Q14 - Baaki Analysis
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.to_period('M')

# Q7: Monthly Trend
plt.figure(figsize=(12,5))
df.groupby('Month')['TotalAmount'].sum().plot(kind='line', marker='o', color='green')
plt.title('Q7: Monthly Sales Trend')
plt.grid(True)
plt.show()

# Q8: Top 10 Products
print("Q8: Top 10 Products by Sales")
print(df.groupby('ProductName')['TotalAmount'].sum().sort_values(ascending=False).head(10))

# Q9: Top 10 Customers
print("\nQ9: Top 10 Customers by Sales")
print(df.groupby('CustomerName')['TotalAmount'].sum().sort_values(ascending=False).head(10))

# Q10: Avg Discount by Category
print("\nQ10: Avg Discount by Category")
print(df.groupby('Category')['Discount'].mean().sort_values(ascending=False))

# Q11: Total Quantity by Category
print("\nQ11: Total Quantity by Category")
print(df.groupby('Category')['Quantity'].sum().sort_values(ascending=False))

# Q12: Discount vs Sales
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='TotalAmount', data=df, alpha=0.5)
plt.title('Q12: Discount vs Total Sales')
plt.show()

# Q13: Top 10 Cities
print("Q13: Top 10 Cities by Sales")
print(df.groupby('City')['TotalAmount'].sum().sort_values(ascending=False).head(10))

# Q14: Top 10 Brands
print("\nQ14: Top 10 Brands by Sales")
print(df.groupby('Brand')['TotalAmount'].sum().sort_values(ascending=False).head(10))
```

```python
plt.figure(figsize=(16,6))

# Q17: Sales by Payment Method - Bar Chart
plt.subplot(1,2,1)
df.groupby('PaymentMethod')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='bar', color='purple')
plt.title('Q17: Sales by Payment Method')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')

# Q18: Heatmap - Top 8 States vs Category
plt.subplot(1,2,2)
pivot = df.pivot_table(values='TotalAmount', index='State', columns='Category', aggfunc='sum').fillna(0)
top_states = df.groupby('State')['TotalAmount'].sum().nlargest(8).index
sns.heatmap(pivot.loc[top_states], cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Q18: Top 8 States vs Category Heatmap')

plt.tight_layout()
plt.show()
```
